```python
import network 
import rarfile
import os

def extract_rar(rar_path: str, extract_to: str):
    """
    Extracts a .rar file to the given folder.

    :param rar_path: Path to the .rar archive
    :param extract_to: Path to the target extraction folder
    """
    if not os.path.exists(extract_to):
        os.makedirs(extract_to)

    with rarfile.RarFile(rar_path) as rf:
        rf.extractall(path=extract_to)
    print(f"Extracted {rar_path} to {extract_to}")


rar_path = r"data/pvs9_video.rar"
extract_to = r"data"

extract_rar(rar_path, extract_to)
```

Testing CUDA enabled training

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from tqdm.notebook import tqdm


print("hello WOrld")
#test
# ---------------------------
# 1. Dummy data
# ---------------------------
num_samples = 50000       # number of dummy images
num_classes = 3         # adapt to your ResNet18Classifier
image_size = (3, 224, 224)  # ResNet18 expects 224x224 RGB

# Random images + random labels
X = torch.randn(num_samples, *image_size)
y = torch.randint(0, num_classes, (num_samples,))

dataset = TensorDataset(X, y)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

# ---------------------------
# 2. Model, Loss, Optimizer
# ---------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = network.ResNet18Classifier(nr_classes=num_classes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# ---------------------------
# 3. Training Loop
# ---------------------------
epochs = 3  

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
    
    for inputs, targets in pbar:
        inputs, targets = inputs.to(device), targets.to(device)

        # Forward
        outputs = model(inputs)
        loss = criterion(outputs, targets)

        # Backward
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        pbar.set_postfix(loss=loss.item())
    
    print(f"Epoch {epoch+1} finished | Avg Loss: {running_loss/len(train_loader):.4f}")
```
